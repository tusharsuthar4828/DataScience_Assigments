```python
import numpy as np
import pandas as pd
```

```python
forest=pd.read_csv('/content/forestfires.csv')
forest.head()
```

```python
forest.pop('month')
forest.head()
```

```python
forest.pop('day')
forest.head()
```

```python
forest.isna().sum()
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
forest['size_category']=label_encoder.fit_transform(forest['size_category'])
```

```python
forest.head()
```

```python
x=forest.iloc[:,0:28]
y=forest.iloc[:,28]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
#Neural Network - ANN
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
#initializing ANN
model=Sequential()
```

```python
#adding input and 1st hidden layer
model.add(Dense(units=10,activation='relu',kernel_initializer='he_uniform',input_dim=28))
#adding 2nd hidden layer
model.add(Dense(units=8,activation='relu',kernel_initializer='he_uniform'))
#adding output layer
model.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
#compile the model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#fit the model
model.fit(x_train,y_train,batch_size=10,epochs=150,validation_split=0.33)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
#evaluating the model
scores = model.evaluate(x_test, y_test)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
